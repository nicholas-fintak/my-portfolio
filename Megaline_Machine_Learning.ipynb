```python
#import necessary libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#read the data
df = pd.read_csv('/datasets/users_behavior.csv')
```

```python
#explore dataset
print(df.info())
```

```python
#RandomForest
#create binary classification target
df.loc[df['is_ultra'] == 1, 'plan_class'] = 1
df.loc[df['is_ultra'] == 0, 'plan_class'] = 0
```

```python
#split data for training and validation
df_train, df_valid = train_test_split(df, test_size=0.25, random_state=12345)
```

```python
#create features and target for training and validation
features_train = df_train.drop(['is_ultra', 'plan_class'], axis=1)
target_train = df_train['plan_class']
features_valid = df_valid.drop(['is_ultra', 'plan_class'], axis=1)
target_valid = df_valid['plan_class']
```

```python
#initialize score, estimator, and best model
best_score_rf = 0
best_est = 0
best_model_rf = None
for est in range(1,101): #hyperparamter range
    model = RandomForestClassifier(random_state=12345, n_estimators=est) #sets number of trees
    model.fit(features_train, target_train) #trains model on training set
    score = model.score(features_valid, target_valid) #calculates accuracy score for validation set
    if score > best_score_rf:
        best_score_rf = score
        best_est = est
        best_model_rf = model #Stores the best model
```

```python
#prints best estimators and accuracy
print("Accuracy of the best model on the validation set (n_estimators= {}): {}".format(best_est, best_score_rf))
```

Initially I set the hyperparameter range (1,11) and achieved an accuracy of 0.7898009950248757 with n_estimators being 10.
I will try again with a hyperparameter range (1,16) to see if I can get a higher score.
After testing with a hyperparameter range of (1,16) the accuracy score went up to 0.7910447761194029 with n_estimators being 12.
I will do one more test with a hyperparameter range of (1,21) to see if the score will increase.
After increasing the hyperparameter range to (1,21) the accuracy remained the same at 0.7910447761194029 and n_estimators also remained the same at 12.
Based upon the feeback received, I tested with a hyperparameter range of (1,101) and was able to achieve an accuracy of 0.80009950248756219 with n_estimators=60. For efficiency, we could lower the range to (1,61).

```python
#prints best details for random forest
print("=== RANDOM FOREST BEST MODEL DETAILS ===")
print(f"Best Model: {best_model_rf}")
print(f"Best Number of Estimators: {best_est}")
print(f"Best Accuracy: {best_score_rf:.4f}")
print(f"Best Accuracy (percentage): {best_score_rf * 100:.2f}%")
```

```python
#DecisionTree
#Initialize score and depth
best_score_dt = 0
best_depth = 0
best_model_dt = None
for depth in range(1,8): #max_depth values
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth) #sets depth
    model.fit(features_train, target_train) #trains model on training set
    score = model.score(features_valid, target_valid) #calculates score
    if score > best_score_dt:
        best_score_dt = score
        best_depth = depth
        best_model_dt = model
```

```python
#prints best depth and score
print("Accuracy of the best model on the validation set (best_depth= {}): {}".format(best_depth, best_score_dt))
```

For the DecisionTree I will use the same three ranges I used for the RandomForest (1,11), (1,16), and (1,21).
Using range(1,11) best_depth=7 and the accuracy is 0.789800990248757.
Using range(1,16) the best_depth and accuracy remained the same as range(1,11).
Using rang(1,21) the best_depth and accuracy were the same as (1,11).
Based upon the tests, the optimal range is (1,8) because the best_depth from the tests was 7 and the model achieved over a 0.75 accuracy.

```python
#prints best details for decision tree
print("=== DECISION TREE BEST MODEL DETAILS ===")
print(f"Best Model: {best_model_dt}")
print(f"Best Score: {best_score_dt:.4f}")
print(f"Best Depth: {best_depth}")
print(f"Best Accuracy (percentage): {best_score_dt * 100:.2f}%")
```

```python
# LogisticRegression with Feature Scaling

#Test different max_iter values with nromal features
best_score_lr = 0
best_max_iter = 0
best_model_lr = None
max_iter_values = [50, 100, 250, 500, 1000]

for max_iter in max_iter_values:
    model = LogisticRegression(random_state=12345, solver='liblinear', max_iter= max_iter)
    model.fit(features_train, target_train)
    score = model.score(features_valid, target_valid)
    if score > best_score_lr:
        best_score_lr = score
        best_max_iter = max_iter
        best_model_lr = model

# Initialize scaler
scaler = StandardScaler()

# Scale the features
features_train_scaled = scaler.fit_transform(features_train)
features_valid_scaled = scaler.transform(features_valid)

# Test different max_iter values with scaled features
best_score_lr_scaled = 0
best_max_iter_scaled = 0
best_model_lr_scaled = None
max_iter_values = [50, 100, 250, 500, 1000]

for max_iter in max_iter_values:
    model = LogisticRegression(random_state=12345, solver='liblinear', max_iter=max_iter)
    model.fit(features_train_scaled, target_train)
    score = model.score(features_valid_scaled, target_valid)
    if score > best_score_lr_scaled:
        best_score_lr_scaled = score
        best_max_iter_scaled = max_iter
        best_model_lr_scaled = model
```

```python
#prints scaled accuracy, scaled max iterations, original accuracy, and the differece in accuracy scores
print(f"SCALED - Best accuracy: {best_score_lr_scaled}")
print(f"SCALED - Best max_iter: {best_max_iter_scaled}")
print(f"ORIGINAL - Best accuracy: {best_score_lr}")
print(f"Improvement: {best_score_lr_scaled - best_score_lr:.4f}")
```

```python
#prints best details for logistic regression
print("=== LOGISTIC REGRESSION BEST SCALED MODEL DETAILS ===")
print(f"Best Scaled Model: {best_model_lr_scaled}")
print(f"Best Scaled Score: {best_score_lr_scaled:.4f}")
print(f"Best Scaled Max Iteration: {best_max_iter_scaled}")
print(f"Best Scaled Accuracy (percentage): {best_score_lr_scaled * 100:.2f}%")
```
