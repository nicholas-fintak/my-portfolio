## 4.1 Regional Platform Analysis

Let's begin by examining platform performance across different regions:

```python
# Function to analyze platform performance by region
def analyze_platform_by_region(data, region_column):
    """
    Analyze platform performance for a specific region
    
    Parameters:
    data: DataFrame with game sales data
    region_column: string, name of the regional sales column (e.g., 'na_sales')
    
    Returns:
    DataFrame with platform analysis results
    """
def analyze_platform_by_region(data, region_column):
    platform_stats = data.groupby('platform')[region_column].agg([
        'count',    # Number of games
        'sum',      # Total regional sales
        'mean',     # Average sales per game
        'median',   # Typical performance
        'std'       # Variability
    ]).round(2)
    
    # Sort by total sales
    platform_stats = platform_stats.sort_values('sum', ascending=False)
    
    return platform_stats
```

## 4.3 ESRB Rating Impact Analysis

Finally, let's examine how ESRB ratings affect sales in each region:

```python
# Function to analyze ESRB rating impact
def analyze_esrb_by_region(data, region_column):
    """
    Analyze ESRB rating impact for a specific region
    
    Parameters:
    data: DataFrame with game sales data
    region_column: string, name of the regional sales column (e.g., 'na_sales')
    
    Returns:
    DataFrame with ESRB rating analysis results
    """
    # Filter out games without ESRB ratings (NaN values)
    data_with_ratings = data.dropna(subset=['rating'])
    
    # Group by ESRB rating and calculate statistics
    rating_stats = data_with_ratings.groupby('rating')[region_column].agg([
        'count',    # Number of games with each rating
        'sum',      # Total regional sales for each rating
        'mean',     # Average sales per game
        'median',   # Typical performance
        'std'       # Variability in sales
    ]).round(2)
    
    # Sort by total sales (highest first)
    rating_stats = rating_stats.sort_values('sum', ascending=False)
    
    return rating_stats
```

```python
# Analyze ESRB impact for each region
na_esrb_analysis = analyze_esrb_by_region(games_relevant, 'na_sales')
print("North America ESRB Rating Performance:")
print(na_esrb_analysis)
eu_esrb_analysis = analyze_esrb_by_region(games_relevant, 'eu_sales')
print("Europe ESRB Rating Performance:")
print(eu_esrb_analysis)
jp_esrb_analysis = analyze_esrb_by_region(games_relevant, 'jp_sales')
print("Japan ESRB Rating Performance:")
print(jp_esrb_analysis)
```

```python
# null hypethesis: The average user ratings of Xbox One and PC platforms are the same.
# alternate hypothesis: The average user ratings of Xbox One and PC platforms are different.

# Set alpha threshold
alpha = 0.05

# Filter data for Xbox One and PC platforms
xbox_ratings = games_relevant[games_relevant['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = games_relevant[games_relevant['platform'] == 'PC']['user_score'].dropna()

# Perform independent t-test
t_statistic, p_value = st.ttest_ind(xbox_ratings, pc_ratings, equal_var=False)

print(f"Xbox One average rating: {xbox_ratings.mean():.3f}")
print(f"PC average rating: {pc_ratings.mean():.3f}")
print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Alpha: {alpha}")

# Interpret results
if p_value < alpha:
    print("Reject null hypothesis: The average ratings are significantly different")
else:
    print("Fail to reject null hypothesis: No significant difference in average ratings")
```

The null hypothesis was formed based upon the instruction at the start of this section.
The alternate was formed to be the opposite of the null hypothesis.
ttest_ind was used because we were looking to compare the ratings between the two platforms.
I believe there was little difference because the Xbox One and PC are both Microsoft platforms.

```python
# Null hypothesis: The average user ratings for Action and Sports genres are the same.
# Alternate hypothesis: he average user ratings for Action and Sports genres are different.

# Filter data for Action and Sports genres
action_ratings = games_relevant[games_relevant['genre'] == 'Action']['user_score'].dropna()
sports_ratings = games_relevant[games_relevant['genre'] == 'Sports']['user_score'].dropna()

alpha = 0.05

# Perform independent t-test
t_statistic, p_value = st.ttest_ind(action_ratings, sports_ratings, equal_var=False)

print(f"Action average rating: {action_ratings.mean():.3f}")
print(f"Sports average rating: {sports_ratings.mean():.3f}")
print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Alpha: {alpha}")

# Interpret results
if p_value < alpha:
    print("Reject null hypothesis: The average ratings are significantly different")
else:
    print("Fail to reject null hypothesis: No significant difference in average ratings")
```
